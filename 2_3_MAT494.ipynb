2.3.1 Joint Probability Distributions

```python
from numpy.random import normal as nom
import matplotlib.pyplot as plt
x, y = nom(size=5000), nom(size=5000)
plt.axis([-5,5,-5,5])
plt.scatter(x, y, s=1)
```

2.3.2 Correlation and Dependence

```python
from numpy import corrcoef as cof
from numpy.random import random as rd
print(cof(x, y), "\n", "-"*30)
print(cof(x, y)[0][1])
print(cof(x, x)[0][1])
print(cof(x, x+100000)[0][1])
print(cof(y, y+100000)[0][1])
print(cof(x, x+10*rd(len(x)))[0][1])
print(cof(x, x+100000*rd(len(x)))[0][1])
```

2.3.3 Random Samples

```python
from numpy.random import gamma, choice
import matplotlib.pyplot as plt

pd = gamma(1, 100, 10000)
print(pd.mean())
ms = [choice(pd, 100).mean() for i in range(20000)]
plt.hist(ms, bins=50)
plt.show()
```
